```python
def strike_price_76(F ,r ,sigma ,  T , t ,delta, option_type = 'call' ):
    if option_type == 'call':
        # result = (F* np.exp( - norm.ppf(delta * np.exp(r*t)) * sigma * np.sqrt(T) - (sigma ** 2 * T) / 2)) 
        result = (F* np.exp( - norm.ppf(delta ) * sigma * np.sqrt(T) - (sigma ** 2 * T) / 2))  # with less error
        return result 
    else:
        delta =  delta + 1
        # result = F* np.exp(  - norm.ppf(delta * np.exp(r*t)) * sigma * np.sqrt(T) - (sigma ** 2 * T) / 2) 
        result = F* np.exp(  - norm.ppf(delta ) * sigma * np.sqrt(T) -(sigma ** 2 * T) / 2) # with less error
        return result


```
